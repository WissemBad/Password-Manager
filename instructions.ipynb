### Introduction
Lors de notre réunion, le client a exprimé son besoin de développer un gestionnaire de mots de passe pour une utilisation professionnelle. L'objectif principal est d'assurer une gestion sécurisée et pratique des identifiants, tout en intégrant des fonctionnalités avancées répondant à des exigences spécifiques.

### Exigences Fonctionnelles
* Cryptage des données de connexion (login et mots de passe):
    * Les mots de passe des données de connexion doivent être protégées par un cryptage à choisir parmi 3 cryptages qui seront implémentés dans l'outil :
        * Un cryptage classique par décalage, dit Code de Césard 
        * Un cryptage basé sur une technologie existante (type AES-256) 
        * Un cryptage basé sur un algorithme que vous proposerez

* Identification de la force d'un mot de passe :
    * Définir une métrique pour la robustesse d'un mot de passe sur une échelle de 4 valeurs (Faible, Moyen, Fort, Excellent), et être capable de l'expliquer à l'utilisateur
    * Calculer la robustesse des mot de passe

* Générateur de mots de passe :
    * Capable de créer des mots de passe complexes selon quatre critères :
        * Longueur souhaitées
        * Types de caractères souhaités (minuscules, majuscules, chiffres, caractères spéciaux)
        * Utilisation de mots du dictionnaire pour rendre le mot de passe plus mémorable
        * Robustesse désirée

* Historique et Gestion des Versions :
    * Sauvegarde automatique des modifications.
    * Accès à l’historique des mots de passe pour restaurer une version précédente.

* Organisation des mots de passe :
    * L'application permettra d'appliquer aux mots de passe des labels
    * Chaque mot de passe peut avoir un ou plusieurs labels

* Menu de gestion en ligne de commande :
    * Une interface en ligne de commande pour gérer le gestionnaire de mots de passe (ajout de données de connexion, suppression, modification...)
    * Permettre d'afficher le mot de passe en clair ou en crypté d'une donnée de connexion
    * La possibilité de chercher dans les données de connexions en fonction soit d'un login, soit d'un label ou soit directement par le mot de passe

### Souhaits Fonctionnels
* Si possible, un stockage sécurisé :
    * Sauvegarder les mots de passe dans un fichier texte, en crypté uniquement (attention à vous assurez que les clés pour crypter soit également fournie et protéger)

* Si possible, charger des mots de passe déjà existant
    * Charger un fichier de mots de passe déjà existant.
